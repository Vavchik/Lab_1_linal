Функция проверки числа на простоту

```python
def is_prime(n: int) -> bool:
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    
    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False
    return True
```

<h2>Task 1</h2>

1. Все палиндромы a < 10^5, такие что a^2 - также палиндром.
2. Все простые числа p < 10^6, все циклические перестановки которых также являются простыми числами.

```python
def get_permutations(n: int) -> List[int]:
    """Возвращает все циклические перестановки числа"""
    s = str(n)
    permutations = []
    for i in range(len(s)):
        permutations.append(int(s[i:] + s[:i]))
    return permutations


def palindromic_squares_and_circular_primes() -> tuple[List[int], List[int]]:
    """
    Возвращает:
        tuple:
            - список всех палиндромов a < 100000, для которых a^2 — палиндром;
            - список всех простых p < 1000000, все циклические перестановки цифр которых просты.
    """
    result = ([], [])

    # Палиндромы a < 10^5, где a^2 — также палиндром
    for a in range(1, 100000):
        if str(a) == str(a)[::-1]:
            if str(a ** 2) == str(a ** 2)[::-1]:
                result[0].append(a)
    
    # Простые числа p < 10^6, все циклические перестановки просты
    for p in range(2, 1000000):
        if is_prime(p):
            permutations = get_permutations(p)
            all_prime = True
            
            # Проверяем все перестановки на простоту
            for perm in permutations:
                if not is_prime(perm):
                    all_prime = False
                    break
            
            if all_prime:
                result[1].append(p)
    
    return result


res = palindromic_squares_and_circular_primes()

print("Палиндромы a < 10^5, где a^2 — палиндром:")
print(res[0])

print("\nКруговые простые числа < 10^6:")
print(res[1])
```

<h2>Task 2</h2>

1. Найти все палиндромы a < 10^5, такие что a^3 - также палиндром.
2. Найди все палиндромические простые числа p <= 10000.

```python
def palindromic_cubes_and_palindromic_primes() -> tuple[List[int], List[int]]:
    """
    Возвращает:
        tuple:
            - список всех палиндромов a < 100000, для которых a^3 — палиндром;
            - список всех простых p <= 10000, которые являются палиндромами.
    """
    result = ([], [])

    # Палиндромы a < 10^5, где a^3 — палиндром
    for a in range(1, 100000):
        if str(a) == str(a)[::-1]:
            if str(a ** 3) == str(a ** 3)[::-1]:
                result[0].append(a)
    
    # Палиндромические простые числа p ≤ 10000
    for p in range(2, 10001):
        if str(p) == str(p)[::-1] and is_prime(p):
            result[1].append(p)
    
    return result


res = palindromic_cubes_and_palindromic_primes()

print("Палиндромы a < 10^5, где a^3 — палиндром:")
print(res[0])

print("\nПалиндромические простые числа ≤ 10000:")
print(res[1])
```

<h2>Task 3</h2>

Первые 100 простых чисел, состоящих только из:
- цифр 1 и 3,
- цифр 1 и 5,
- цифр 1 и 7,
- цифр 1 и 9.

```python
def primes_with_two_digits() -> Dict[str, List[int]]:
    """
    Возвращает словарь вида:
        {
            '13': [список первых 100 простых из {1,3}],
            '15': [список первых 100 простых из {1,5}],
            '17': [список первых 100 простых из {1,7}],
            '19': [список первых 100 простых из {1,9}]
        }
    """
    result = {'13': [], 
              '15' : [], 
              '17': [],
              '19': []}
    
    pairs = [('1', '3'), ('1', '5'), ('1', '7'), ('1', '9')]
    
    for d1, d2 in pairs:
        if is_prime(int(d1)):
            result[d1 + d2].append(int(d1))
        if is_prime(int(d2)):
            result[d1 + d2].append(int(d2))
        cur_length = 2
        while len(result[d1 + d2]) <= 100:

            def generate_numbers(nums: List[str], pref: str = '') -> None:
                if pref:
                    if len(pref) == cur_length:
                        if is_prime(int(pref)):
                            result[d1 + d2].append(int(pref))
                        return
                for num in nums:
                    generate_numbers(nums, pref + num)
            
            generate_numbers([d1, d2])
            cur_length += 1

        result[d1 + d2] = result[d1 + d2][:100]
    
    return result


result = primes_with_two_digits()
for key, nums in result.items():
    print(f"Простые числа, состоящие только из {key}:")
    print(nums)
```

*Меньше всего простых чисел, стоятоящих из цифр 1 и 5. Причина тому, скорее всего, цифра 5, ведь если число заканчивается на 5, то оно точно не простое (тк кратно 5)*

<h2>Task 4</h2>

Первые 1000 пар-близнецов (p, p + 2).

```python
def twin_primes_analysis(limit_pairs: int = 1000) -> Tuple[List[Tuple[int, int]], List[float]]:
    """
    Возвращает:
        - список первых `limit_pairs` пар близнецов (p, p+2);
        - список значений отношения pi_2(n) / pi(n) для n, соответствующих последним элементам каждой пары,
        где pi_2(n) — количество пар близнецов <= n, pi(n) — количество простых <= n.
    """
    pairs = []
    ratios = []
    
    pi = 0
    pi_2 = 0
    n = 2
    
    while len(pairs) < limit_pairs:
        if is_prime(n):
            pi += 1
            # Проверяем, является ли пара (n-2, n) близнецами
            if is_prime(n + 2):
                pi_2 += 1
                pairs.append((n, n + 2))
                ratio = pi_2 / pi
                ratios.append(ratio)
        n += 1
    
    return pairs, ratios


twin_pairs, ratios = twin_primes_analysis()
print("Первые 5 и последние 5 пар простых-близнецов:")
for i, (p1, p2) in enumerate(twin_pairs[:5] + twin_pairs[995:]):
    print(f"{i+1}. ({p1}, {p2})")

print("\nЗначения pi_2(n) / pi(n) для каждой 100-й пары:")
for i, ratio in enumerate(ratios[::100]):
    print(f"{i + 1}. {ratio}")
print(f"Отношение для последней пары: {ratios[-1]:.6f}")
```

*При росте n отношение пар близнецов к общему числу простых при росте n уменьшается*

<h2>Task 5</h2>

Вычисление разложение числа n! + 1 на простые множители для 2 <= n <= 50.

```python
def factorial_plus_one_factors() -> Dict[int, Dict[int, int]]:
    """
    Возвращает словарь вида:
    { n: {простой делитель: степень, ...}, ... }
    для n от 2 до 50, где ключ - n, значение - разложение n! + 1 на простые множители.
    """
    result = {}
    
    for n in range(2, 51):
        # Вычисляем n! + 1
        number = factorial(n) + 1
        # Факторизуем с помощью sympy
        factors = factorint(number)
        result[n] = factors
    
    return result
        
        
# Вычисляем все разложения
factorizations = factorial_plus_one_factors()

# Максимальное количество различных простых делителей
max_distinct_primes = 0
n_with_max_primes = []

# Случаи с "большими" простыми множителями (> 10^6)
large_prime_cases = set()

for n in range(2, 51):
    factors = factorizations[n]
    num_distinct_primes = len(factors)
    
    # Обновляем максимум по количеству различных простых делителей
    if num_distinct_primes > max_distinct_primes:
        max_distinct_primes = num_distinct_primes
        n_with_max_primes = [n]
    elif num_distinct_primes == max_distinct_primes:
        n_with_max_primes.append(n)
    
    # Проверяем на наличие больших простых множителей
    primes = list(factors.keys())
    large_primes = [p for p in primes if p > 10**6]
    
    if large_primes:
        for prime in large_primes:
            large_prime_cases.add(prime)

print(f'Число с максимальным количесвтом простых делителей: {n_with_max_primes}')
print(max_distinct_primes)
print(f'Большие простые делители: {large_prime_cases}')
```

<h2>Task 6</h2>

Два способа вычисления функции Эйлера f(n):
1. Прямой перебор
2. Через разложение на простые множители

```python
import time
from typing import List
from sympy.functions.combinatorial.numbers import totient


def gcd(a: int, b: int) -> int:
    """Вычисление НОД двух чисел."""
    while b:
        a, b = b, a % b
    return a


def eulier_phi_direct(n: int) -> int:
    """Вычисление (n) прямым перебором."""
    count = 0
    for k in range(1, n + 1):
        if gcd(n, k) == 1:
            count += 1
    return count


def eulier_phi_prime_factors(n: int) -> int:
    """Вычисление функции Эйлера f(n) через разложение на простые множители."""
    result = n
    p = 2
    while p * p <= n:
        if n % p == 0:
            while n % p == 0:
                n //= p
            result -= result // p
        p += 1
    if n > 1:
        result -= result // n
    return result


def compare_eulier_phi_methods(test_values: List[int]) -> dict:
    """Сравнение двух методов вычисления функции Эйлера f(n) для n от 1 до limit."""
    result = {}
    for value in test_values:
        for n in range(1, value + 1):
            start_direct = time.time()
            result_direct = eulier_phi_direct(n)
            time_direct = time.time() - start_direct
            
            start_prime = time.time()
            result_prime = eulier_phi_prime_factors(n)
            time_prime = time.time() - start_prime

            start_sympy = time.time()
            result_sympy = totient(n)
            time_sympy = time.time() - start_sympy
        
        result[n] = {
                'direct': (result_direct, time_direct),
                'prime_factors': (result_prime, time_prime),
                'sympy': (result_sympy, time_sympy)
            }
        
    return result


res = compare_eulier_phi_methods([10, 100, 1000])

for n, methods in res.items():
    result_direct, time_direct = methods['direct']
    result_prime_factors, time_prime_factors = methods['prime_factors']
    result_sympy, time_sympy = methods['sympy']
    
    assert result_direct == result_prime_factors == result_sympy, f"Несовпадение результатов для n={n}"
    print(f"n={n}: f(n)={result_direct}, Прямой перебор: {time_direct} с, Через простые множители: {time_prime_factors} с, SymPy: {time_sympy} с")
```
