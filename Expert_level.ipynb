<h2>Task 1</h2>

Проверить, что лексикографический порядок и порядок по степени - отношения порядка на мономах.

**1. Лексикографический порядок (lex)**

x1^a1 ... xn^an > x1^b1 ... xn^bn, если существует i: a1 = b1, ..., a(i-1) = b(i-1), ai > bi

**Транзитивность:**  
Пусть A <= B и B <= C.  
Рассмотрим первую позицию i, где ai ≠ ci.  
Если для всех j < i aj = bj = cj, то в позиции i:  
ai <= bi и bi <= ci ⇒ ai <= ci.  
Если ai < ci, то A < C.  
Если ai = ci, то переходим дальше.  
В итоге либо A = C, либо A < C

**Линейность:** 
Любые два монома сравнимы

**Мультипликативность:**


✅ **lex — отношение порядка (линейное)**


## 2. Порядок по степени (griex - graded lexicographic order)

**Определение:**  
- Сначала сравниваем общую степень |a| = sum(ai) и |b| = sum(bi)
- Если |a| > |b|, то A > B
- Если |a| = |b|, то сравниваем лексикографически (как в lex)

**Транзитивность:**  
Пусть A <= B и B <= C.  
Тогда |a| <= |b| <= |c| => |a| <= |c|.  
- Если |a| < |c|, то A < C
- Если |a| = |c|, то |a| = |b| = |c|, и в этом случае порядок — lex, а lex транзитивен => A <= C

**Линейность:** любые два монома сравнимы

✅ **griex — отношение порядка (линейное)**


**Утверждение ВЕРНО.**

<h2>Task 5</h2>

Доказать или опровергнуть, что любой базис Грёбнера можно привести к редуцированному.

**Определение**

Редуцированный базис Грёбнера удовлетворяет:

1. LC(g_i) = 1 для всех i (старший коэффициент равен 1)
2. Ни один моном полинома (g_i) не делится ни на какой LM(g_j) при j != i (редуцированность относительно остальных)

**Алгоритм приведения**

Пусть G — произвольный базис Грёбнера

*1. Нормировка старших коэффициентов*  

Заменить каждый g_i на (1 / LC(g_i))* g_i  
Это не меняет идеала LT(G), так как LM(g_i) сохраняется, а LC(g_i) становится 1

*2. Редукция каждого полинома*

Для каждого g_i выполнить:  

g_i <- g_i - (m / LT(h)) * h (те пока в g_i есть моном, который делится на LM(h), где h принадлежит G, переписываем в g_i остатки от деления)

Поскольку G — базис Грёбнера, LM(g_i) не изменится (так как LM(g_i) не делится на LM(g_j) при j != i, иначе g_i можно было бы убрать из базиса)

Редукция убирает из g_i все мономы, делящиеся на LM(g_j) j != i

После этого g_i редуцирован

*3. Удаление лишних полиномов*

Если LM(g_i) делится на LM(g_j) для i != j, то g_i можно удалить из базиса  
После удаления повторить редукцию

<h2>Task 6</h2>

Реализовать алгоритм Бухберга

```python
from itertools import combinations

class Polynomial:
    """Простой класс для работы с полиномами"""
    
    def __init__(self, terms):
        """
        terms: словарь {моном: коэффициент}
        моном - кортеж (степень_x, степень_y, ...)
        пример: x² + 2xy -> {(2,0): 1, (1,1): 2}
        """
        self.terms = {monom: coeff for monom, coeff in terms.items() if coeff != 0}

        if self.terms:  # Сортировка
            sorted_terms = sorted(self.terms.keys(), reverse=True)
            self.terms = {monom: self.terms[monom] for monom in sorted_terms}    
    
    def leading_monomial(self):
        """Старший моном LM"""
        if not self.terms:
            return None
        return next(iter(self.terms.keys()))
    
    def leading_coeff(self):
        """Старший коэффициент LC"""
        if not self.terms:
            return 0
        return next(iter(self.terms.values()))
    
    def leading_term(self):
        """Старший терм (кортеж из LM и LC)"""
        return (self.leading_monomial(), self.leading_coeff())
    
    def is_zero(self):
        """Проверка на ноль"""
        return len(self.terms) == 0
    
    def __add__(self, other):
        """Сложение полиномов"""
        result = self.terms.copy()
        for monom, coeff in other.terms.items():
            result[monom] = result.get(monom, 0) + coeff
        return Polynomial({m: c for m, c in result.items() if c != 0})
    
    def __sub__(self, other):
        """Вычитание полиномов"""
        result = self.terms.copy()
        for monom, coeff in other.terms.items():
            result[monom] = result.get(monom, 0) - coeff
        return Polynomial({m: c for m, c in result.items() if c != 0})
    
    def __mul__(self, other):
        """Умножение на полином"""
        if isinstance(other, (int, float)):  # Если это число
            return Polynomial({monom: coeff * other for monom, coeff in self.terms.items()})
        
        # Умножение полиномов
        result = {}
        for monom1, coeff1 in self.terms.items():
            for monom2, coeff2 in other.terms.items():
                new_monom = tuple(monom1[i] + monom2[i] for i in range(len(monom1)))
                result[new_monom] = result.get(new_monom, 0) + coeff1 * coeff2
        return Polynomial(result)
    
    def __rmul__(self, other):
        """Умножение справа на число"""
        if isinstance(other, (int, float)):
            return self * other
        return NotImplemented
    
    def __str__(self):
        if self.is_zero():
            return "0"
        
        parts = []
        for monom, coeff in self.terms.items():
            if sum(monom) == 0:  # константа
                parts.append(f"{coeff}")
            else:
                vars_str = ""
                for i, power in enumerate(monom):
                    if power > 0:
                        vars_str += f"x{i+1}" + (f"^{power}" if power > 1 else "")
                if coeff == 1:
                    parts.append(vars_str)
                elif coeff == -1:
                    parts.append(f"-{vars_str}")
                else:
                    parts.append(f"{coeff}{vars_str}")
        
        result = " + ".join(parts).replace("+ -", "- ")
        return result

def S_polynomial(f, g):
    """Вычисление S-полинома для f и g"""
    if f.is_zero() or g.is_zero():
        return Polynomial({})
    
    lm_f, lc_f = f.leading_term()
    lm_g, lc_g = g.leading_term()

    # Находим НОК мономов
    lcm = tuple(max(lm_f[i], lm_g[i]) for i in range(len(lm_f)))
    
    # Вычисляем множители
    print(lm_f, lm_g)
    factor_f = tuple(lcm[i] - lm_f[i] for i in range(len(lm_f)))
    factor_g = tuple(lcm[i] - lm_g[i] for i in range(len(lm_g)))
    
    # Создаем полиномы-множители
    poly_f = Polynomial({factor_f: 1.0})
    poly_g = Polynomial({factor_g: 1.0})
    
    # S = (LCM/LT(f)) * f - (LCM/LT(g)) * g
    part1 = (1/lc_f) * (poly_f * f)
    part2 = (1/lc_g) * (poly_g * g)
    return part1 - part2

def divide(p, divisors):
    """Деление полинома p на список divisors"""
    r = Polynomial({})
    
    while not p.is_zero():
        lm_p, lc_p = p.leading_term()
        divided = False
        
        for g in divisors:
            if g.is_zero():
                continue
                
            lm_g, lc_g = g.leading_term()
            
            # Проверяем, делится ли старший моном p на старший моном g
            if all(lm_p[i] >= lm_g[i] for i in range(len(lm_p))):
                # Вычисляем множитель
                factor = tuple(lm_p[i] - lm_g[i] for i in range(len(lm_p)))
                factor = Polynomial({factor: 1.0})
                
                # p = p - (LT(p)/LT(g)) * g
                p = p - (lc_p / lc_g) * factor * g
                divided = True
                break
        
        if not divided:
            # Добавляем к остатку
            r = r + Polynomial({lm_p: lc_p})
            # Убираем из p
            new_terms = p.terms.copy()
            del new_terms[lm_p]
            p = Polynomial(new_terms)
    
    return r

def buchberger(F):
    """Алгоритм Бухбергера для нахождения базиса Грёбнера"""
    G = [f for f in F if not f.is_zero()]
    pairs = list(combinations(range(len(G)), 2))
    
    while pairs:  # Повторяем, пока N(S(f_i, f_j)) = 0
        i, j = pairs.pop(0)
        f, g = G[i], G[j]
        
        # Вычисляем S-полином
        S = S_polynomial(f, g)
        
        # Делим на текущий базис (редуцирование)
        remainder = divide(S, G)
        
        # Если остаток не ноль, добавляем r в G
        if not remainder.is_zero():
            G.append(remainder)
            # Добавляем новые пары с новым полиномом
            new_index = len(G) - 1
            for k in range(new_index):
                pairs.append((k, new_index))
    
    return G
```

```python
print("Один полином")
f1 = Polynomial({(2,): 1, (1,): -1}) 
basis = buchberger([f1])
print([str(g) for g in basis])

print("Два одинаковых полинома")
f1 = Polynomial({(1,): 1, (0,): -1}) 
f2 = Polynomial({(1,): 1, (0,): -1}) 
basis = buchberger([f1, f2])
print([str(g) for g in basis])

print("Полином и его множитель")
f1 = Polynomial({(2,): 1, (0,): -1}) 
f2 = Polynomial({(1,): 1, (0,): -1}) 
basis = buchberger([f1, f2])
print([str(g) for g in basis])

print("Два независимых полинома")
f1 = Polynomial({(2, 0): 1})
f2 = Polynomial({(0,1): 1}) 
basis = buchberger([f1, f2])
print([str(g) for g in basis])

print("Три полинома")
f1 = Polynomial({(1, 0): 1, (0, 0): -1})
f2 = Polynomial({(0,1): 1, (0, 0): -2}) 
f3 = Polynomial({(1,1): 1})  
basis = buchberger([f1, f2, f3])
print([str(g) for g in basis])

print("Проверка деления")
f = Polynomial({(2,): 1, (1,): 2, (0,): 1})
g = Polynomial({(1,): 1, (0,): 1})
remainder = divide(f, [g])
print(f"Остаток: {remainder}")

print("S-полином")
f1 = Polynomial({(2,): 1})  
f2 = Polynomial({(1,): 1})  
S = S_polynomial(f1, f2)
print(f"S(f1,f2) = {S}")
```

<h2>Task 7</h2>

Проверить, принадлежит ли полином f идеалу, порожденному базисом G

```python
def is_in_ideal(f, G):
    """

    Args:
        f: Polynomial - проверяемый полином
        G: list[Polynomial] - редуцированный базис Грёбнера (старшие коэффициенты = 1
    Returns:
        bool: True если f принадлежит идеалу, иначе False
    """
    remainder = divide(f, G)  # редуцирование
    
    return remainder.is_zero()
```

```python
print("Полином в идеале")
G = [Polynomial({(1,): 1, (0,): -1})]
f = Polynomial({(2,): 1, (0,): -1})  
print(f"G = {[str(g) for g in G]}")
print(f"f = {f}")
print(f"f в идеале - {is_in_ideal(f, G)}")

print("Полином не в идеале")
G = [Polynomial({(1,): 1, (0,): -1})]
f = Polynomial({(2,): 1, (1,): 1})
print(f"G = {[str(g) for g in G]}")
print(f"f = {f}")
print(f"f в идеале - {is_in_ideal(f, G)}")

print("Нулевой полином")
G = [Polynomial({(1,): 1})]
f = Polynomial({})         
print(f"G = {[str(g) for g in G]}")
print(f"f = {f}")
print(f"f в идеале - {is_in_ideal(f, G)}")

print("Идеал с двумя образующими")
G = [
    Polynomial({(2,): 1, (0,): -1}),
    Polynomial({(1,): 1, (0,): -1})
]
f = Polynomial({(3,): 1, (0,): -1})
print(f"G = {[str(g) for g in G]}")
print(f"f = {f}")
print(f"f в идеале - {is_in_ideal(f, G)}")

print("Многомерный случай")
G = [
    Polynomial({(1,0): 1, (0,0): -1}),
    Polynomial({(0,1): 1, (0,0): -2})
]
f = Polynomial({(1,1): 1, (1,0): -2, (0,1): -1, (0,0): 2})
print(f"G = {[str(g) for g in G]}")
print(f"f = {f}")
print(f"f в идеале - {is_in_ideal(f, G)}")
```
